```python
from create_local_data import get_train_data
import numpy as np
```

```python
test = get_train_data()
```

```python
test[0].shape
```

```python
test[1].shape
```

```python
test_new = np.concatenate((test[0], test[1]), axis=-1)
```

```python
for i in range(test_new.shape[1]):
    mean = test_new[:,i].mean()
    std = test_new[:,i].std()
    test_new[:,i] -= mean
    test_new[:,i] /= std
```

```python
test_new.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(test_new)
test2 = pca.transform(test_new)
```
